```python
import gymnasium as gym
import numpy as np
```

__NO ACTIONS__

```python
# NOTE: render_mode=human renders it on pygame window
env = gym.make('LunarLander-v3', continuous=True, render_mode='human')

# NOTE: This will instead make a viedo folder
# env = gym.wrappers.RecordVideo(env, 'video', video_length=1_000)

obs, _ = env.reset()

# NOTE: this will create five scenes of the spaceship
for i in range(5):
    while True:
        # NOTE: this has no action because it is all zeros, it is random 
        action = np.zeros(env.action_space.shape[0])

        obs,_,terminated,truncated, _ = env.step(action)

        if terminated or truncated:
            obs,_ = env.reset()
            break
    
env.close()
```

__ACTIONS__

```python
from IPython.display import display, clear_output
from PIL import Image
import time
# NOTE: we can use rgb_array to create instance of the video
env = gym.make('LunarLander-v3', continuous=True, render_mode="rgb_array")

obs, _ = env.reset()

# NOTE: this will create five scenes of the spaceship
for i in range(5):
    while True:
        # NOTE: now actions are sampled randomly at each scene, this can be visualized in the pygame window,
        #       In this case they are called thrust
        action = env.action_space.sample()
        obs,_,terminated,truncated, _ = env.step(action)

        # NOTE: following makes the output be visible as oputput cell
        frame = env.render()
        clear_output(wait=True)
        display(Image.fromarray(frame))
        time.sleep(0.03)

        if terminated or truncated:
            obs,_ = env.reset()
            break
    
env.close()
```

Actions now are random, no policy is applied and in order for us to nail our goal we need to effectively take actions that will make the spaceship land correctly.

The simulation demonstrates an agent, spaceship, takling actions in an environment, space. This is are the basic components of a reinforcement learning setup

__GOAL:__ learn the policy which decides whioch actions to take
